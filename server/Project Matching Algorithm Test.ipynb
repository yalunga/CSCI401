```python
def runAlgorithm(students, rankings, projects):
    NUM_RANKED = 5
    P_MAX = (((NUM_RANKED-1+1) * (NUM_RANKED-1)) / 2) + 1
    best_assignment = {}
    best_sat = 0
    for i in range(30):
        print("Running iteration ", i)
        project_p_sum = {}
        project_n = {}
        project_popularity = {}
        assignments = {}
        total_max_spots = get_total_max_spots(projects)
        eliminated = []
        for j, row in rankings.iterrows():
            student_id = row['SID']
            project_id = row['ID']
            project_sat_score = (((NUM_RANKED-row['Ranking']+1) * (NUM_RANKED-row['Ranking'])) / 2) + 1
            project_p_sum[project_id] = project_p_sum[project_id] + project_sat_score if project_id in project_p_sum else project_sat_score
            project_n[project_id] = project_n[project_id] + 1 if project_id in project_n else 1
        if projects.shape[0] < NUM_RANKED:
            NUM_RANKED = projects.shape[0]
        for j, row in projects.iterrows():
            first = (2 * project_p_sum[j/2+1]) / (project_n[j/2+1] * P_MAX) if j/2+1 in project_p_sum else 0
            popularity = (first + (project_n[j/2+1] / row['Maximum Students'])) / 3 if j/2+1 in project_n else 0
            project_popularity[j/2+1] = popularity
        sorted_projects = get_sorted_projects(project_popularity)
        
        # Assign initial
        unassigned_students = [x for x in range(students.shape[0])]
        random.shuffle(unassigned_students)
        students_list = [x for x in unassigned_students]
        for student in students_list:
            for choice in range(NUM_RANKED):
                student_rows = rankings[rankings['SID'] == student]
                student_ranking_row = student_rows[student_rows['Ranking'] == choice+1]
                project = student_ranking_row['ID'].values[0]
                if project not in eliminated:
                    if project not in assignments:
                        assignments[project] = [student]
                        unassigned_students.remove(student)
                        break
                    elif len(assignments[project]) < projects['Maximum Students'][2*(project-1)] and student not in assignments[project]:
                        assignments[project] = assignments[project] + [student]
                        unassigned_students.remove(student)
                        break
        
        # Eliminate projects
        # Add people from unranked pool
        for j in range(projects.shape[0]-1, 0, -1):
            if ((j not in assignments) or len(assignments[j]) < projects[projects['Project #'] == j]['Minimum Students'].values[0]) and (total_max_spots - projects[projects['Project #'] == j]['Maximum Students'].values[0]) >= students.shape[0]:
                if assignments[project]:
                    eliminated_students = assignments[project]
                    for s in eliminated_students:
                        unassigned_students.append(s)
                eliminated.append(project)
        
        # Bump projects
        random.shuffle(unassigned_students)
        students_list = [x for x in unassigned_students]
        for student in students_list:
            if student in unassigned_students and bumpHelper(student, 0, rankings, eliminated, projects, assignments):
                unassigned_students.remove(student)
        
        # Iteration total SAT score
        totalSat = get_total_sat(assignments, projects, NUM_RANKED, P_MAX, rankings)
        num_assigned = projects.shape[0] - len(eliminated)
        algoSatScore = totalSat / num_assigned
        print("Iteration: ", i, ", Satisfaction: ", algoSatScore)
        print(assignments)
        
        # Place unassigned
        if len(unassigned_students) > 0:
            assignments['Unassigned'] = []
            for s in unassigned_students:
                assignments['Unassigned'].append(s)
        
        if algoSatScore > best_sat:
            best_sat = algoSatScore
            best_assignment = assignments
        print("BEST SAT SCORE: ", best_sat)
    return best_assignment
```

```python
def get_total_sat(assignments, projects, NUM_RANKED, P_MAX, rankings):
    total = 0
    for i, row in projects.iterrows():
        max_score = P_MAX * row['Maximum Students']
        score = 0
        if row['Project #'] in assignments:
            for student in assignments[row['Project #']]:
                student_rows = rankings[rankings['SID'] == student]
                student_ranking_row = student_rows[student_rows['ID'] == row['Project #']]
                ranking = student_ranking_row['Ranking'].values[0]
                score += (((NUM_RANKED-ranking+1) * (NUM_RANKED-ranking)) / 2) + 1
            
        total += score / max_score
    return total
```

```python
def bumpHelper(student, level, rankings, eliminated, projects, assignments):
    if level > 3:
        return False
    for i in range(5):
        student_rows = rankings[rankings['SID'] == student]
        student_ranking_row = student_rows[student_rows['Ranking'] == i+1]
        project = student_ranking_row['ID'].values[0]
        if project not in eliminated and projects['Maximum Students'][2*(project-1)]:
            if project not in assignments:
                assignments[project] = [student]
                return True
            elif student not in assignments[project]:
                assignments[project] = assignments[project] + [student]
                return True
    student_rows = rankings[rankings['SID'] == student]
    student_ranking_row = student_rows[student_rows['Ranking'] == 1]
    first_choice = student_ranking_row['ID'].values[0]
    
    idx = random.randint(0, len(assignments[first_choice])-1)
    displaced = assignments[first_choice][idx]
    if bumpHelper(displaced, level+1, rankings, eliminated, projects, assignments):
        if student not in assignments[first_choice] and first_choice not in eliminated:
            if len(assignments[first_choice]) < projects['Maximum Students'][2*(first_choice-1)] and len(assignments[first_choice]) + 1 > projects['Minimum Students'][2*(first_choice-1)]:
                assignments[first_choice].remove(displaced)
                assignments[first_choice].append(student)
                
            
        
    return True
```

```python
def get_total_max_spots(projects):
    total = 0
    for i, row in projects.iterrows():
        total += row['Maximum Students']
    return total
```

```python
import operator
def get_sorted_projects(project_popularity):
    sorted_projects = []
    for i in range(len(project_popularity)):
        max_key = max(project_popularity.items(), key=operator.itemgetter(1))[0]
        sorted_projects.append(max_key)
        project_popularity[max_key] = -100
    return sorted_projects
```
